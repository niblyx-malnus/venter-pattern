# Chat-GPT Generated Diagrams

## Generate ID Client-Side

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure
fig, ax = plt.subplots()

# Create extended rectangles for Client and Server
client_rect = patches.Rectangle((0.1, 0.3), 0.2, 0.4, edgecolor='blue', facecolor='none')
server_rect = patches.Rectangle((0.7, 0.3), 0.2, 0.4, edgecolor='green', facecolor='none')

# Add the rectangles to the Axes
ax.add_patch(client_rect)
ax.add_patch(server_rect)

# Add labels for Client and Server
ax.text(0.15, 0.75, 'Client', ha='center', va='center', fontsize=12, color='blue')
ax.text(0.75, 0.75, 'Server', ha='center', va='center', fontsize=12, color='green')

# Add a section inside the Client rectangle for ID generation
id_gen_circle = patches.Circle((0.2, 0.5), 0.08, edgecolor='blue', facecolor='none')
ax.add_patch(id_gen_circle)
ax.text(0.2, 0.5, 'Generate ID', ha='center', va='center', fontsize=8, color='blue')

# Create an arrow for the request from Client to Server
ax.annotate('',
            xy=(0.7, 0.5), xycoords='data',
            xytext=(0.28, 0.5), textcoords='data',
            arrowprops=dict(arrowstyle="->", color='black'))

# Label the request arrow
ax.text(0.5, 0.55, 'Request (Data, ID)', ha='center', va='center', fontsize=8)

# Create an arrow for the acknowledgment from Server to Client
ax.annotate('',
            xy=(0.3, 0.4), xycoords='data',
            xytext=(0.7, 0.4), textcoords='data',
            arrowprops=dict(arrowstyle="->", color='black'))

# Label the acknowledgment arrow
ax.text(0.5, 0.35, 'Ack', ha='center', va='center', fontsize=8)

# Set limits and remove axis
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Plotting
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Generate ID Client-Side")

plt.savefig('images/generate_ID_client_side.png', format='png')

plt.show()
```

## Ideal Request-Response Cycle

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure
fig, ax = plt.subplots()

# Create extended rectangles for Client and Server
client_rect = patches.Rectangle((0.1, 0.3), 0.2, 0.4, edgecolor='blue', facecolor='none')
server_rect = patches.Rectangle((0.7, 0.3), 0.2, 0.4, edgecolor='green', facecolor='none')

# Add the rectangles to the Axes
ax.add_patch(client_rect)
ax.add_patch(server_rect)

# Add labels for Client and Server
ax.text(0.15, 0.75, 'Client', ha='center', va='center', fontsize=12, color='blue')
ax.text(0.75, 0.75, 'Server', ha='center', va='center', fontsize=12, color='green')

# Adjust the position of the 'Generate ID' section to be horizontally centered in the Server rectangle
generate_id_y_position = 0.4
generate_id_x_position = 0.8  # Horizontal center of the server rectangle
id_gen_circle_server = patches.Circle((generate_id_x_position, generate_id_y_position), 0.08, edgecolor='green', facecolor='none')
ax.add_patch(id_gen_circle_server)
ax.text(generate_id_x_position, generate_id_y_position, 'Generate ID', ha='center', va='center', fontsize=8, color='green')

# Create an arrow for the request from Client to Server
ax.annotate('',
            xy=(0.7, 0.5), xycoords='data',
            xytext=(0.3, 0.5), textcoords='data',
            arrowprops=dict(arrowstyle="->", color='black'))

# Label the request arrow
ax.text(0.5, 0.55, 'Request (Data)', ha='center', va='center', fontsize=8)

# Create a horizontal arrow for the response from Server to Client, starting at the left edge of the Generate ID circle
response_arrow_start_x = generate_id_x_position - 0.08  # Left edge of the circle
ax.annotate('',
            xy=(0.3, generate_id_y_position), xycoords='data',
            xytext=(response_arrow_start_x, generate_id_y_position), textcoords='data',
            arrowprops=dict(arrowstyle="->", color='black'))

# Label the response arrow
ax.text(0.5, generate_id_y_position - 0.05, 'Response (ID)', ha='center', va='center', fontsize=8)

# Set limits and remove axis
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Plotting
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Ideal ID Generating Request-Response Cycle")

plt.savefig('images/ideal_ID_request_response.png', format='png')

plt.show()
```

## Poke Forwarding with a Nack

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure
fig, ax = plt.subplots()

# Create rectangles for Client, Server A, and Server B
client_rect = patches.Rectangle((0.1, 0.1), 0.2, 0.8, edgecolor='blue', facecolor='none')
serverA_rect = patches.Rectangle((0.4, 0.1), 0.2, 0.8, edgecolor='red', facecolor='none')
serverB_rect = patches.Rectangle((0.7, 0.1), 0.2, 0.8, edgecolor='green', facecolor='none')

# Add the rectangles to the Axes
ax.add_patch(client_rect)
ax.add_patch(serverA_rect)
ax.add_patch(serverB_rect)

# Add labels for Client, Server A, and Server B
ax.text(0.2, 0.95, 'Client', ha='center', va='center', fontsize=12, color='blue')
ax.text(0.5, 0.95, 'Server A', ha='center', va='center', fontsize=12, color='red')
ax.text(0.8, 0.95, 'Server B', ha='center', va='center', fontsize=12, color='green')

# Create an arrow for the request from Client to Server A
ax.annotate('',
            xy=(0.4, 0.7), xycoords='data',
            xytext=(0.3, 0.7), textcoords='data',
            arrowprops=dict(arrowstyle="->", color='black'))
# Label the request arrow
ax.text(0.35, 0.75, 'Request', ha='center', va='center', fontsize=10)

# Create an arrow for the forwarded request from Server A to Server B
ax.annotate('',
            xy=(0.7, 0.6), xycoords='data',
            xytext=(0.6, 0.6), textcoords='data',
            arrowprops=dict(arrowstyle="->", color='black'))
# Label the forwarded request arrow
ax.text(0.65, 0.65, 'Forwarded Request', ha='center', va='center', fontsize=10)

# Create an arrow for the acknowledgment from Server A to Client
# Align the right side of the arrow with the left side of Server A
ax.annotate('',
            xy=(0.3, 0.5), xycoords='data',
            xytext=(0.4, 0.5), textcoords='data',
            arrowprops=dict(arrowstyle="->", color='black'))
# Label the acknowledgment arrow
ax.text(0.35, 0.45, 'Ack to Client', ha='center', va='center', fontsize=10)

# Create an arrow for the Nack from Server B to Server A
ax.annotate('',
            xy=(0.6, 0.4), xycoords='data',
            xytext=(0.7, 0.4), textcoords='data',
            arrowprops=dict(arrowstyle="->", color='black'))
# Label the Nack arrow
ax.text(0.65, 0.35, 'Nack (Crash)', ha='center', va='center', fontsize=10)

# Set limits and remove axis
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')


# Plotting
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Poke Forwarding with a Nack")

plt.savefig('images/poke_forwarding_nack.png', format='png')

plt.show()
```

## Ideal Poke Forwarding

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure
fig, ax = plt.subplots()

# Create rectangles for Client, Server A, and Server B
client_rect = patches.Rectangle((0.1, 0.1), 0.2, 0.8, edgecolor='blue', facecolor='none')
serverA_rect = patches.Rectangle((0.4, 0.1), 0.2, 0.8, edgecolor='red', facecolor='none')
serverB_rect = patches.Rectangle((0.7, 0.1), 0.2, 0.8, edgecolor='green', facecolor='none')

# Add the rectangles to the Axes
ax.add_patch(client_rect)
ax.add_patch(serverA_rect)
ax.add_patch(serverB_rect)

# Add labels for Client, Server A, and Server B
ax.text(0.2, 0.95, 'Client', ha='center', va='center', fontsize=12, color='blue')
ax.text(0.5, 0.95, 'Server A', ha='center', va='center', fontsize=12, color='red')
ax.text(0.8, 0.95, 'Server B', ha='center', va='center', fontsize=12, color='green')

# Create an arrow for the request from Client to Server A
ax.annotate('', xy=(0.4, 0.7), xycoords='data', xytext=(0.3, 0.7), textcoords='data', arrowprops=dict(arrowstyle="->", color='black'))
ax.text(0.35, 0.75, 'Request to Server A', ha='center', va='center', fontsize=10)

# Create a dotted arrow for the report crash in Server A, going to the Client
ax.annotate('', xy=(0.3, 0.6), xycoords='data', xytext=(0.4, 0.6), textcoords='data', arrowprops=dict(arrowstyle="->", color='black', linestyle=':'))
ax.text(0.35, 0.55, 'Report Crash in Server A', ha='center', va='center', fontsize=10)

# Create an arrow for the forwarded request from Server A to Server B
ax.annotate('', xy=(0.7, 0.5), xycoords='data', xytext=(0.6, 0.5), textcoords='data', arrowprops=dict(arrowstyle="->", color='black'))
ax.text(0.65, 0.45, 'Forward Request to Server B', ha='center', va='center', fontsize=10)

# Create an arrow for the response or crash from Server B to Server A
ax.annotate('', xy=(0.6, 0.4), xycoords='data', xytext=(0.7, 0.4), textcoords='data', arrowprops=dict(arrowstyle="->", color='black'))
ax.text(0.65, 0.35, 'Response or Crash', ha='center', va='center', fontsize=10)

# Create an arrow for the forward response or crash from Server A to Client, starting at the left side of Server A
ax.annotate('', xy=(0.3, 0.3), xycoords='data', xytext=(0.4, 0.3), textcoords='data', arrowprops=dict(arrowstyle="->", color='black'))
ax.text(0.35, 0.25, 'Forward Response or Crash', ha='center', va='center', fontsize=10)

# Set limits and remove axis
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Plotting
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Ideal Poke Forwarding")

plt.savefig('images/ideal_poke_forwarding.png', format='png')

plt.show()
```

## Venter Pattern

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a new figure
fig, ax = plt.subplots()

# Create rectangles for Program A and Program B
programA_rect = patches.Rectangle((0.1, 0.2), 0.3, 0.6, edgecolor='blue', facecolor='none')
programB_rect = patches.Rectangle((0.6, 0.2), 0.3, 0.6, edgecolor='green', facecolor='none')

# Add the rectangles to the Axes
ax.add_patch(programA_rect)
ax.add_patch(programB_rect)

# Add labels for Program A and Program B
ax.text(0.25, 0.85, 'Program A', ha='center', va='center', fontsize=12, color='blue')
ax.text(0.75, 0.85, 'Program B', ha='center', va='center', fontsize=12, color='green')

# Create an arrow for %watch from Program A to Program B
ax.annotate('', xy=(0.6, 0.7), xycoords='data', xytext=(0.4, 0.7), textcoords='data', arrowprops=dict(arrowstyle="->", color='black'))
ax.text(0.5, 0.75, '%watch /vent/[vent-id]', ha='center', va='center', fontsize=10)

# Create an arrow for vent-request from Program A to Program B
ax.annotate('', xy=(0.6, 0.6), xycoords='data', xytext=(0.4, 0.6), textcoords='data', arrowprops=dict(arrowstyle="->", color='black'))
ax.text(0.5, 0.65, 'Vent-Request (vent-id, action)', ha='center', va='center', fontsize=10)

# Internal Computation in Program B
computation_circle = patches.Circle((0.75, 0.45), 0.1, edgecolor='green', facecolor='none')
ax.add_patch(computation_circle)
ax.text(0.75, 0.45, 'Computation', ha='center', va='center', fontsize=8, color='green')

# Dotted arrows for asynchronous computation in Program B
# Outward dotted arrow above the inward dotted arrow
ax.annotate('', xy=(0.95, 0.48), xycoords='data', xytext=(0.84, 0.48), textcoords='data', arrowprops=dict(arrowstyle="->", color='black', linestyle=':'))
ax.annotate('', xy=(0.95, 0.42), xycoords='data', xytext=(0.84, 0.42), textcoords='data', arrowprops=dict(arrowstyle="<-", color='black', linestyle=':'))

# Create an arrow for %fact with a Vent from Program B to Program A, vertically centered with the computation circle
ax.annotate('', xy=(0.4, 0.45), xycoords='data', xytext=(0.65, 0.45), textcoords='data', arrowprops=dict(arrowstyle="->", color='black'))
ax.text(0.525, 0.4, '%fact (vent)', ha='center', va='center', fontsize=10)

# Create an arrow for %kick from Program B to Program A, starting at the left side of Program B
ax.annotate('', xy=(0.4, 0.3), xycoords='data', xytext=(0.6, 0.3), textcoords='data', arrowprops=dict(arrowstyle="->", color='black'))
ax.text(0.5, 0.25, '%kick', ha='center', va='center', fontsize=10)

# Set limits and remove axis
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Plotting
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Venter Pattern")

plt.savefig('images/venter_pattern.png', format='png')

plt.show()
```
